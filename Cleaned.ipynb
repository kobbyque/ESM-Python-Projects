```python
#importing pandas
import pandas as pd

#Upload the CSV file and read it into a DataFrame
from google.colab import files
uploaded = files.upload()

# Assuming the file name is 'movies.csv'
df = pd.read_csv('glassdoor.csv')


# Inspect the first few rows of the DataFrame
print("First five rows of the dataset:")
print(df.head())

# Get a summary of the DataFrame
print("\nDataFrame summary:")
print(df.info())

df['company_name'].value_counts()

# Check how many values are missing in each column
df.isnull().sum().sort_values(ascending=False)

# See how many unique values each column has
df.nunique().sort_values()

df = df.drop(columns=['company_name'])
# Adjust based on your findings

# Count how many missing values are in each column
df.isnull().sum().sort_values(ascending=False)

# Drop rows where the main review text is missing
df = df.dropna(subset=['summary'])

df.info()

# Remove any exact duplicate rows
df = df.drop_duplicates()

# Save the cleaned DataFrame to a new CSV
df.to_csv('glassdoor_cleaned.csv', index=False)
```
